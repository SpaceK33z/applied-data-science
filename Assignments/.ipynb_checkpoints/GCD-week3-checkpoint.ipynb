```python
import pandas as pd

def strip(text):
    try:
        return text.strip()
    except AttributeError:
        return text

data = pd.read_csv('./data/grain.csv', sep=';', encoding='utf-8')


data.head()
```

# Trim whitespace

```python
columns = list(data.columns)
columns.remove('Hectares')

for c in columns:
    data[c] = data[c].str.strip()

# Test to see if the stripping worked.
data.groupby('Status of deal').count()
```

```python
a = data['Projected investment']
a.replace(['US\$', ' .*$'], '', regex=True, inplace=True)
        
data.groupby('Projected investment').count()
```
